```python
import feedparser

def get_article_urls(rss_feed_url):
    feed = feedparser.parse(rss_feed_url)
    urls = [entry.link for entry in feed.entries]
    return urls
```

```python
from newspaper import Article

def extract_article_info(url):
    article = Article(url)
    article.download()
    article.parse()
    
    title = article.title
    author = article.authors
    publish_date = article.publish_date
    content = article.text
    
    return {
        'title': title,
        'author': author,
        'publish_date': publish_date,
        'content': content
    }
```

```python
rss_feed_url = 'https://www.sciencenews.org/topic/artificial-intelligence/feed'
urls = get_article_urls(rss_feed_url)
    
for url in urls:
    info = extract_article_info(url)
    print(f"Title: {info['title']}")
    print(f"Author: {info['author']}")
    print(f"Publish Date: {info['publish_date']}")
    print(f"Content: {info['content'][:500]}...")
    print("\n" + "-"*80 + "\n")
```
